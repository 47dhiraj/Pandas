# To apply your own or another library’s functions to Pandas objects, you should be aware of the three important methods. The methods have been discussed below. The appropriate method to use depends on whether your function expects to operate on an entire DataFrame, row- or column-wise, or element wise.

# Table wise Function Application: pipe()
# Row or Column Wise Function Application: apply()
# Element wise Function Application: applymap()

#  1> Table-wise Function Application --> is performed on the whole DataFrame.

# For example, add a value 2 to all the elements in the DataFrame.

```python
import numpy as np
import pandas as pd

def adder(ele1,ele2):            # adder function
   return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
```

```python
df
```

```python
df.pipe(adder,2)
```

# 2> Row or Column Wise Function Application --> using the apply() method

 # apply() method can take inbuilt pandas method , custom function & lamba expressions

# Column Wise : 

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
```

```python
df
```

```python
df.apply(np.mean)                # By Default apply method le paramter axis = 0 lincha i.e Colum wise mean nikleko xa
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.apply(lambda x: x.max() - x.min())                        # Using Lambda Expression in apply() method
```

```python
df
```

```python
df.apply(np.mean)
```

# Row Wise :

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.apply(np.mean,axis=1)                  # Row wise apply method lai use garna ko lagi  axis = 1 lekhna parxa
```

```python
df
```

```python
df.apply(np.mean)
```

# Element Wise Function Application

```python
# Not all functions can be vectorized (neither the NumPy arrays which return another array nor any value), the methods applymap() on DataFrame and analogously map() on Series accept any Python function taking a single value and returning a single value.
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
```

```python
df
```

```python
df.apply(np.mean)
```

```python
df['col1'].map(lambda x:x*100)
```
