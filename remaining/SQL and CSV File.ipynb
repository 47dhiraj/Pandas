# Appending the data to the table

```python
df1 = pd.DataFrame({'name' : ['Shankar', 'Siva']})
df1.to_sql('users', con=engine, if_exists='append')
engine.execute("SELECT * FROM users").fetchall()
```

# MySQL real databse sanga connection garera, Database ma dataframe store garera, paxi fetch Database ko data lai fetch pani gareko

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

user = 'root'
passw = ''
host =  '127.0.0.1'  # either localhost or ip e.g. '172.17.0.2' or hostname address 
port = 3306 
database = 'panda_Test'

engine = create_engine('mysql+pymysql://' + user + ':' + passw + '@' + host + ':' + str(port) + '/' + database , echo=False)

df = pd.DataFrame({'name' : ['Ram', 'Hari', 'Shyam']})

df.to_sql('users', con=engine, if_exists = 'replace', index=False)

#df.to_sql(name='Users', con=engine, if_exists = 'replace', index=False)

engine.execute("SELECT * FROM users").fetchall()
```
