```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)  # merging two dataframes(movies & ratings)

ratings
```

```python
user_ratings1 = ratings.pivot_table(index=['userId'], columns=['title'], values='rating')

user_ratings1
```

```python
# Let's Remove Movies which have less than 10 users who rated it. and fill remaining NaN with 0
user_ratings1 = user_ratings1.dropna(thresh=5,axis=1).fillna(0)        # dropna(thresh=5, axis=1) --> yesko kaam vaneko,  jun column (i.e movie) ma 5 vanda kam users le rating deko xa.. testo column (axis=1 vannale column bujinxa) lai drop gar vaneko  & again,  fillna(0) ko kaam vaneko column drop gari sake paxi jati column bachxa .. bacheko column ma pani NaN value tw huna saki halxa.. so tyo NaN value lai 0 le replace garni vaneko..  

user_ratings1
```
