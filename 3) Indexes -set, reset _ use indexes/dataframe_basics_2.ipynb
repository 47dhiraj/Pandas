```python
# For Reading csv File
# df = pd.read_csv('filename.csv')
# OR
# df = pd.read_csv('data.csv', encoding='utf-8')
# df

# df.head()              # First five rows dekhaucha
# df.head(50)

# df.tail()              # last Five rows dekhaucha
# df.tail(30)

# print(df)

# Note : csv file load garera dataframe create garne bela mai indexing as column garna pani sakincha & encoding pani provide garna sakincha
df = pd.read_csv('data.csv', index_col='Title',encoding='utf-8')       

# Get the number of rows and columns 
all_rows = len(df.axes[0]) 
all_cols = len(df.axes[1]) 
print(all_rows)
print(all_cols)
```

```python
# Note : By Default Jupyter notebook will display only first 5 rows and last 5 rows of dataframe.. obviously kahile kai tmle aafno dataframe ko sabai rows herna khoji rako huncha.. so testo bela yo use garne
# khas ma yo csv file ko lagi jupyter notebook ko setting milayeko jastai ho
# SYNTAX ==> pd.set_option('display.max_rows', all_rows_number)
df = pd.set_option('display.max_rows', all_rows)
```

```python
# Note : By Default Jupyter notebook will display only first 20 columns .. obviously tmle aafno dataframe ko sabai columns herna khoji rako huncha.. so testo bela yo use garne
# khas ma yo csv file ko lagi jupyter notebook ko setting milayeko jastai ho
# SYNTAX ==> pd.set_option('display.max_columns', all_columns_number)
df = pd.set_option('display.max_columns', all_cols)
```
