### `Writing .csv file with unnecesary dataframe index, is a bad way of writing csv file`

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Dhiraj\Desktop\pandas-for-interview\datasets\countries of the world.csv")
```

```python
df
```

### Avoid writing `unnecessary dataframe index to .csv file`

### `1st way:`  use `index=False`

```python
# NOTE: The index=False parameter specifies that we don't want to write the index/row-index of dataframe to the csv file.

# NOTE: By default, index is also written to .csv, but only in rare case we need it. So,index=False is written to avoid index.

df.to_csv('properly_written_countries_of_the_world.csv', index=False)
```

```python
# reading the dataframe

df2 = pd.read_csv(r"properly_written_countries_of_the_world.csv")

df2
```

### `2nd way:` use `index_col=[0]` while `reading unproperly written .csv file`

```python
df_read_properly = pd.read_csv(r"mistake_written_countries_of_the_world.csv", index_col=[0])

df_read_properly
```
