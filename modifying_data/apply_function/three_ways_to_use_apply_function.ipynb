```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("NYC_Jobs.csv")

data = data[['Job ID','Civil Service Title','Agency','Posting Type','Job Category','Salary Range From','Salary Range To']]

data.head()
```

## apply can be used with dataframe and series object

###  Using a built-in function (like  apply(np.sqrt) )

#### NOTE: yedi numeric built in function lagaune ho vani, only numeric colum/series object ma lagauna, whole dataframe ma tyo numeric function lagauna khojyo vani attribute error falcha

```python
# you can apply a pre-defined function to it
data['Salary Range From'].apply(np.sqrt)
```

```python
# # if u use double brackets [[ ]] then, pandas will return u a dtaframe, not the series object

# data[['Salary Range From','Salary Range To']]
# type(data[['Salary Range From','Salary Range To']])    # yesko data type dataframe ho


# yedi dataframe ma apply() function use gare vani, return vako data type pani dataframe object nai huncha
data[['Salary Range From','Salary Range To']].apply(np.sqrt)
# type(data[['Salary Range From','Salary Range To']].apply(np.sqrt))
```

#### For getting average values 

```python
data[['Salary Range From','Salary Range To']].apply(np.mean, axis=1)         # axis=0 vannale, column wise & axis=1 vannale, row wise(i.e accross each row)
# type(data[['Salary Range From','Salary Range To']].apply(np.mean, axis=1))
```

```python
# so if you want to add/assign returned/calculated seires object to original dataframe as new column u can do like following
data['Avg salary'] = data[['Salary Range From','Salary Range To']].apply(np.mean, axis=1)
```

### Create your own Custon function

```python
# you can create your own custom function & apply it to the dataframe
def capitalize_position(title):
    title_lower = title.lower()
    title_final = title_lower.capitalize()
    return title_final


data['Civil Service Title'].apply(capitalize_position)
```

```python
# again if you would need to add/assign the returned result to the original dataframe, u can do like following
data['Civil Service Title'] = data['Civil Service Title'].apply(capitalize_position)
```

### Create a Lambda Function

```python
# NOTE:  x:  vannale, chai lambda function ma parameter ho
# NOTE:  x.tolist(), chai lambda function definition ho

# applying lambda function to whole dataframe
data.apply(lambda x: x.tolist(), axis=1)          # axis=1 vannale row-wise(i.e accross each row)
# type(data.apply(lambda x: x.tolist(), axis=1))    # data type chai series object ho

# # applying lambda function to series object
# data['Agency'].apply(lambda x: x.lower().capitalize())
```

```python
# expand attribute will assign each element of the list to a different column
data.apply(lambda x: x.tolist(), axis=1, result_type='expand')
```

### Broadcasting the resulting lists to match the shape of the original data

### result_type='broadcast' use garera list batw generate vako dataframe ma column pani rakhna milcha

```python
# broadcast would do that and keep the column names
# see that in expand the column names are newly created to go from 0 to 7
data.apply(lambda x: x.tolist(), axis=1, result_type='broadcast')
```
