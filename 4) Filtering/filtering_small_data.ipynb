```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# filter cha column anusar gareko cha, and output dida chai row or record output dincha
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')      # since filter is python keyword so variable banauda ekdam khyal garni.. python ko keyword vako variable banauna paidana so hamile filt vanni banako
filt
```

```python
# filt      # yo filt variable ma boolean value aai rako cha i.e True or False
df[filt]       # filter garera lyaeko value lai as dataframe herni tarika
```

```python
# df.loc[] method ko faida vaneko .. filter garera aako dataframe lai pani aajai filter sajilari garna sakincha ... but note that loc[] method ko lagi first parameter row nai huna parcha & second parameter column nai huna parcha
# df.loc[filt, 'email']           # filter vayera aako data ma pani aajai filter gareko cha i.e email column matra chaiyo vanna lai

# Note : ydi filter condition ko vanda tykka or just opposite value chaiyeko cha vani  ~  (i.e tilda) use garna sakincha 
# df.loc[~filt, 'email']      
```

```python
# Alternative in one line
# df[(df['last'] == 'Schafer') | (df['first'] == 'John')]        # filter cha column anusar gareko cha, and output dida chai row or record output dincha
# OR Best Alternativ i.e filtering using df.loc[]
# df.loc[(df['last'] == 'Schafer') | (df['first'] == 'John'), 'email']

# Note : ydi filter condition ko vanda tykka or just opposite value chaiyeko cha vani  ~  (i.e tilda) use garna sakincha 
df.loc[~(df['last'] == 'Schafer') | (df['first'] == 'John'), 'email']
```
