```python
# Note : ConvertedComp vanni column ma jasko salary 7000 vanda thulo cha rows or records filter out gareko
high_salary = (df['ConvertedComp'] > 7000)
# high_salary           # yesma just Boolean output matra  huncha  i.e True or False
# df[high_salary]         # actual filer garera aayeko records or rows herna paryo vani yesto garincha 


# df.loc[] method ko faida vaneko .. filter garera aako dataframe lai pani aajai filter sajilari garna sakincha ... but note that loc[] method ko lagi first parameter rows or records nai huna parcha & second parameter cha vani columns nai huna parcha
# yo talako high_salary chai records or rows ho & MainBranch, ConvertedComp, Country, LanguageWorkedWith chai columns ho
# df.loc[high_salary]
df.loc[high_salary, ['MainBranch', 'ConvertedComp', 'Country', 'LanguageWorkedWith']]         # filter vayera aako data ma pani aajai filter gareko cha i.e output records ma MainBranch, ConvertedComp, Country, LanguageWorkedWith vanni column ko information matra chaiyo vanna lai
```

```python
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']          # filter yedi dherai kura ko basis ma garnu cha vani yesari list banayera garda sajilo huncha.. i.e filter statement complex or long dekhidaina
filt = df['Country'].isin(countries)           # isin() vanni pandas ko method ho ...  # j ko basis ma filter garni ho tesko list haru isin() method vitra as a parameter ko rup ma paas garincha

# Without using loc[] method
# df[(df['Country'].isin(countries))]        # yesma & or | or help batw filtering condition thapna chai milcha but chuttai yo column ko matra information de vanera specify garna chai mildaina  ,, yedi testo garni ho vani talako loc[] method use garna parcha.. which is considerd as advance than this.


# Using loc[] Method
# df.loc[filt]                 # Mathi ko countries vako recorrds matra filtero outa garera dincha
# df.loc[filt, 'Country']           # Country matra disply gar vaneko..
df.loc[filt, ['Country']]       # yedi mathi kai output lai alli ramo sanga display garnu cha vani tyo column name ko lai big brackeg le cover garney
```

```python
# Filter condition for string (i.e alphabet characters or word) using str method of python i.e str.contains() method 

# LanguageWorkedWith vanni column ko lagi 'Python' string value vako wala records nikal vaneko & note: na=False garnu ko karan vaneko yedi LanguageWorkedWith vanni column ko value NaN cha vani teti bela teslai ignore garna ko lagi na=False lekhna parcha.. haina vani filter contdion right huda hudai pani error aaucha.. feri ka batw error aayo tha pauna nikai time lagna sakcha 
filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
```

```python
# filt
df.loc[filt]
```

```python
# df.loc[filt]
# yedi LanguageWorkedWith column ko value matra filter gareko output ma dekhauna cha vani
# df.loc[filt, 'LanguageWorkedWith']          # yesle ra talako satatement le same output dincha but talako le chai alli ramro sanga a dataframe  ko rup ma output display garcha
df.loc[filt, ['LanguageWorkedWith']]           # chyakka ramro sanga as dataframe tarikale output herna cha vani yesari column name lai big bracket le cover garna parcha
```
